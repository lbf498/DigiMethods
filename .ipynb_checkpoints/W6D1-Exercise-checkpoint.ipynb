**Exercise 6.1.1.1** Read in the dataset "nchs_death_causes.csv" using `pandas` methods.

```python
import pandas as pd
```

```python
us_deaths_2016 = pd.read_csv('nchs_death_causes.csv')

us_deaths_2016
```

**Exercise 6.1.1.2** Make a `DataFrame` that is a subset of the data, containing only one cause of death and one state.

```python
cause_death = pd.DataFrame(us_deaths_2016)

cause_death
```

```python
test_cause_Kidney = cause_death [(cause_death['Cause Name'].isin(['Kidney disease'])) & (cause_death['State'].isin(['Alaska']))]
test_cause_Kidney
```

```python
cause_Kidney = cause_death [cause_death['Cause Name'].isin(['Kidney disease'])]
```

```python
Kidney_Alaska = cause_Kidney [cause_Kidney['State'].isin(['Alaska'])]

Kidney_Alaska
```

**Exercise 6.1.1.3** Plot a line graph that shows the development of the death rate over time. Pay attention to whether your plot is intuitive.

```python
import matplotlib.pyplot as plt
```

```python
values_death_rate = Kidney_Alaska['Age-adjusted Death Rate'].tolist ()
#values_death_rate
values_year = Kidney_Alaska['Year'].tolist()

#OBS husk at omvende begge!!
values_year.reverse()
values_death_rate.reverse()

plt.figure(figsize=(10, 7), dpi= 120)

plt.plot(values_death_rate)

plt.title("Death by Kidney disease in Alaska")
plt.xlabel("Year")
plt.ylabel("Age-adjusted Death Rate")
plt.xticks(range(0,len(values_year)), values_year, rotation=90)
```

```python
import os

values_death_rate = Kidney_Alaska['Age-adjusted Death Rate'].tolist ()
values_year = Kidney_Alaska['Year'].tolist()

#OBS husk at omvende begge!!
values_year.reverse()
values_death_rate.reverse()

plt.figure(figsize=(10, 7), dpi= 120)

plt.plot(values_death_rate)

plt.title("Death by Kidney disease in Alaska")
plt.xlabel("Year")
plt.ylabel("Age-adjusted Death Rate")
plt.xticks(range(0,len(values_year)), values_year, rotation=90)

# Her gemmer vi filen: 
file_dir = os.getcwd()

if '/' in file_dir:
    print("I am a Mac.")
    plt.savefig(file_dir + "/Kidney_Death_Alaska.png")
else:
    print("I am a Windows PC.")
    plt.savefig(file_dir + "\\Kidney_Death_Alaska.png")
```
